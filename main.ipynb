```python
# How many total movies are there in the given data ?
movie_count=netflix_dataset.isnull().sum()["Rating"]
movie_count
```

```python
# Aim is to segrregate the movies IDs from customer columns into the new colums movie_ID
```

```python
movie_id=None
movie_np=[]
for cust_id in netflix_dataset["Cust_ID"]:
  if ':' in cust_id:
    movie_id=int(cust_id.replace(":",""))
  movie_np.append(movie_id)
```

```python
netflix_dataset["Movie_ID"]=movie_np
```

# **Pre filtering manually **

```python
# movie filter
# remove the movies with less number of ratings
data_movie_summary=netflix_dataset.groupby("Movie_ID")["Rating"].count()
data_movie_summary
```

```python
plt.boxplot(data_movie_summary)
```

```python
#setting up benchmark
movie_benchmark=round(data_movie_summary.quantile(0.60))
movie_benchmark
```

```python
(60/100)*4499
```

```python
(40/100)*4499
```

```python
Rejected_movies=data_movie_summary[data_movie_summary<movie_benchmark].index
Rejected_movies
```

```python
# Customer Filter
# removing the customers who have rated less number of movies
```

```python
data_cust_summary=netflix_dataset.groupby("Cust_ID")["Rating"].count()
```

```python
data_cust_summary
```

```python
cust_benchmark=round(data_cust_summary.quantile(0.6))
cust_benchmark
```

```python
Rejected_customers=data_cust_summary[data_cust_summary<cust_benchmark].index
Rejected_customers
```

```python
#Removing the rejected movies and rejected customers from the data
```

```python
netflix_dataset=netflix_dataset[~netflix_dataset["Cust_ID"].isin(Rejected_customers)]
# ~ is exclusion or negation operator which will exclude the resulted data from the actual data
```

```python
netflix_dataset=netflix_dataset[~netflix_dataset["Movie_ID"].isin(Rejected_movies)]
```

```python
netflix_dataset
```
